```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df = pd.read_csv('sms.tsv', sep='\t', header=None, names=['label', 'message'])
print(df.shape)
df.head()
```

```python
df['label_num'] = df['label'].map({'ham': 0, 'spam': 1})
print(df['label'].value_counts())
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df['message'], df['label_num'], test_size=0.2, random_state=42
)
```

```python
tfidf = TfidfVectorizer(stop_words='english', max_features=3000)
X_train_tf = tfidf.fit_transform(X_train)
X_test_tf = tfidf.transform(X_test)

model = MultinomialNB()
model.fit(X_train_tf, y_train)
```

```python
y_pred = model.predict(X_test_tf)
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Ham', 'Spam'],
            yticklabels=['Ham', 'Spam'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('SMS Spam Classifier — Confusion Matrix')
plt.savefig('confusion_matrix.png', dpi=150)
plt.show()
```

```python
samples = ["Win a free iPhone now!!!", "Hey, are we still meeting at 5?"]
samples_tf = tfidf.transform(samples)
predictions = model.predict(samples_tf)
for msg, pred in zip(samples, predictions):
    print(f"{'SPAM' if pred == 1 else 'HAM'}: {msg}")
```
